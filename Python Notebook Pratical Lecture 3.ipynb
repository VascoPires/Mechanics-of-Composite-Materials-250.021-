### Examples of Tensors in Python

```python
# Here, we are defining, 3, random, tensors

# Define two generic second-order tensors (3x3)
A = np.random.rand(3, 3)
B = np.random.rand(3, 3)

# Optionally define a fourth-order tensor (3x3x3x3)
Z = np.random.rand(3, 3, 3, 3)

print("Tensor A:\n", A)
print("Tensor B:\n", B)
```

```python
print("Tensor Z:\n", Z)
```

```python
# Some operations:

# Matrix multiplication using einsum
# This computes the matrix product of A and B using the Einstein summation convention
C = np.einsum('ik,kj->ij', A, B)
print(C)

# Verify that this matches standard matrix multiplication using the normal matrix product np.dot
print('-------')
print(np.dot(A, B))
```

```python
# Double contraction

# The result here is a scalar value, which is the sum of the products of the corresponding elements of A and B
result = np.einsum('ij,ij->', A, B)
print("Double contraction result:", result)


# Let's do the same thing using a nested loop
sum = 0
for i in range(3):
    for j in range(3): 
        sum += A[i][j] * B[i][j]
print("Double contraction result:", sum)
```
