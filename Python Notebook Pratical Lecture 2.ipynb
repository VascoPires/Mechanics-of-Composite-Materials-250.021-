### 3.5) Boolean Expressions
What goes into the `{conditions that the loop continues}` area? It's simply a boolean expression.

**What's a boolean expression?** It's essentially a question with a yes or no answer or a true or false situation. For example, if you want to check if your favorite color matches your friend's, you would write:

My favorite color == my friend's favorite color

And if it's true, you both have a match! Now, if you're older than your sibling, you'd lay it out like this:

My age > my sibling's age

And if it's true, you've got the upper hand. But if you mistakenly claim:

My age < my sibling's age

And your sibling is, in fact, younger, that's a false statement. When this statements are for example in a loop, the code will keep running while the statement used as condition holds true.

### 3.6) `If`, `else` & `elif` (Conditional Statements)

Understanding conditional statements in Python is very important for any programmer. They're the same as making decisions in real life—based. Under certain conditions you may act differently.

Imagine you're deciding what to wear based on the weather. In Python, it'd look something like this:

````python
weather = "sunny"

if weather == "sunny":
    print("Wear sunglasses!")  # If it's sunny, wear sunglasses
elif weather == "rainy":
    print("Grab an umbrella!")  # If it's rainy, bring an umbrella
else:
    print("Enjoy your day!")  # If it's neither sunny nor rainy, just enjoy your day
````

In this example, Python checks the weather. If it's sunny, it tells you to wear sunglasses. If it's rainy, it advises you to grab an umbrella. And if it's neither sunny nor rainy, it simply tells you to enjoy your day.

Now, let's break down the syntax a bit:

- if starts the conditional statement.
- elif stands for "else if" and is used for additional conditions.
- else is the catch-all option if none of the previous conditions are met.


### 3.7) Lists

Similar to the previous sections, understanding lists in Python is necessary for most programs. Lists are basically ways to handle and store data.

Let's explore a simple example where we have a list of numbers and we want to perform some operations on them:

````python
numbers = [1, 2, 3, 4, 5]

# Iterate through each number in the list
for number in numbers:
    # Check if the number is even
    if number % 2 == 0:
        print(f"{number} is even.")
    # If not even, it must be odd
    else:
        print(f"{number} is odd.")
````

In this example, we have a list of numbers [1, 2, 3, 4, 5]. We iterate through each number in the list using a for loop. Within the loop, we use an if statement to check if the number is even or odd. Here's the breakdown:

- The % operator gives the remainder of the division.
- If the remainder is 0, the number is even; otherwise, it's odd.
- if and else statements help categorize the numbers accordingly.

## 4) Rule of Mixtures (RoM)

Now that some of the basics are covered, we will write a code to compute some homogenized properties for the rule of mixtures. The key formulas for the rule of mixtures are the following:


### 4.1) RoM Parallel (Voigt model)

The mechanical properties can be obtained by:

$$E_{l}=\xi E_{l}^{(f)}+(1-\xi)E^{(m)}$$
$$\nu_{lq}=\xi\nu_{lq}^{(f)}+(1-\xi)\nu^{(m)}$$

And the stress and strain are obtained by

$$\varepsilon_{total}=\varepsilon^{(f)}=\varepsilon^{(m)}$$
$$\sigma_{total}=\xi \sigma^{(f)}+(1-\xi) \sigma^{(m)}$$

### 4.2) RoM Parallel (Voigt model)

Similarly for the Voigt model, the mechanical properties are obtained by:

$$E_{q}=E_{t}=\frac{1}{\frac{\xi}{E_{q}^{(\mathrm{f})}}+\frac{(1-\xi)}{E^{(\mathrm{m})}}}$$
$$G_{lq}=\frac{1}{\frac{\xi}{G_{lq}^{({f})}}+\frac{(1-\xi)}{G^{(\mathrm{m})}}}\quad G_{qt}=\frac{1}{\frac{\xi}{G_{qt}^{({f})}}+\frac{(1-\xi)}{G^{({m})}}}$$

And the stress and strain are:
$$\sigma_{total}=\sigma^{(f)}=\sigma^{(m)}$$
$$\varepsilon_{total}=\xi \varepsilon^{(f)}+(1-\xi) \varepsilon^{(m)}$$
